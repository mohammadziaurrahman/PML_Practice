```python
from pandas.api.types import is_numeric_dtype
```

```python
for x in df3.columns:
    if is_numeric_dtype(df3[x]):
        continue
    else:
        uni = df3[x].unique().tolist()
        if df3[x].nunique() == 2:
             df3[x] = df3[x].replace(uni,[1,2])
        elif df3[x].nunique() == 3:
            df3[x] = df3[x].replace(uni,[1,2,3])
        elif df3[x].nunique() == 4:
            df3[x] = df3[x].replace(uni,[1,2,3,4])
df3.head()
```

# One Hot Encoding

```python
df6.head()
```

```python
dummy = pd.get_dummies(df6['sex'], drop_first=True).astype(int)
```

```python
dummy.head()
```

```python
df6 = df6.drop('sex',axis = 1)
df6 = pd.concat([df6,dummy],axis =1)
df6.head()
```

```python
# using for loop to encode all the necessary columns
for x in df6.columns:
    if is_numeric_dtype(df6[x]):
        continue
    else:
        one_encode = pd.get_dummies(df6[x], drop_first=True, prefix = x).astype(int)
        df6.drop(df6[[x]],axis=1, inplace = True)
        df6 = pd.concat([df6,one_encode],axis=1)

df6.head()
```
