```python
# Eucleadian distance
from sklearn.metrics.pairwise import euclidean_distances

x = np.array([[1,2,5]])
y = np.array([[5,8,6]])

eu = euclidean_distances(x,y)
print(eu)
```

```python
x = np.array([[4,1]])
y = np.array([[3,0]])
eu = euclidean_distances(x,y)
eu
```

```python
eu = np.sqrt((x[0,0]-y[0,0])**2 + (x[0,1]-y[0,1])**2)
eu
```

```python
# Manhaattan Distance
from sklearn.metrics.pairwise import manhattan_distances

x = np.array([[1,2]])
y = np.array([[3,4]])

mnhtn = manhattan_distances(x,y)
mnhtn
```

```python
mnhtn = np.abs(x[0,0]-y[0,0]) + np.abs(x[0,1]-y[0,1])
mnhtn
```

```python
# Minkowski Distance
x = np.array([[2,3]])
y = np.array([[5,7]])

from sklearn.metrics import DistanceMetric
dist = DistanceMetric.get_metric('minkowski',p=1) # p = 1 for manhattan distance
dist.pairwise(x,y)
```

```python
# Minkowski Distance with p = 3
mnkdis = ((abs(x[0,0] - y[0,0])**3 + abs(x[0,1] - y[0,1])**3))**(1/3)
mnkdis
```

```python
x = df3['Marketing Spend']

#Compute Q1 (25th percentile) and Q3 (75th percentile)

Q1 = np.percentile(x, 25)  # First quartile
Q3 = np.percentile(x, 75)  # Third quartile

# Compute Median
median_x = np.median(x)

# Compute Interquartile Range (IQR)
IQR = Q3 - Q1

x_robust_scaled = (x - median_x) / IQR

x_robust_scaled
```
